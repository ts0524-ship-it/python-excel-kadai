```python
import openpyxl
```

```python
Wrkbook = openpyxl.Workbook()
```

```python
ws = Wrkbook.active
ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'
ws['F4'] = 'No.'
ws['F5'] = '日付'
ws['G4'] = '0001'
```

```python
ws.column_dimensions['B'].width = 30
ws.column_dimensions['F'].width = 10
ws.column_dimensions['G'].width = 15
```

```python
from datetime import datetime
today = datetime.today()
ws['G5'] = today.strftime('%Y/%m/%d')
```

```python
herder = ['','商品名', '数量', '単価', '金額']
ws.append([])  # 空行を追加
ws.append([])  # 空行を追加
ws.append(herder)  # ヘッダー行を追加
```

```python
data =  [
['','商品A', 2, 10000, 20000],
['','商品B', 1, 15000, 15000]
]
```

```python
total = 0  # 合計金額の初期値
# 商品情報を表に追加
```

```python
for row in data:
    ws.append(row)
    total += row[4]
```

```python
ws['E13'] = 35000
```

```python
ws.append([])  # 空行を追加
ws.append(['', '小計', '','', total])
ws.append(['', '消費税', '','', int(total * 0.1)])  # 消費税率は10%と仮定
ws.append(['', '税込合計', '','', total + int(total * 0.1)])
```

```python
backup_filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"
Wrkbook.save(backup_filename)
```
