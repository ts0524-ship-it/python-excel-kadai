```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
```

```python
# 2つのExcelファイルを読み込む
df_2022 = pd.read_excel("2022_年間売上表.xlsx", sheet_name="Sheet1")
df_2023 = pd.read_excel("2023_年間売上表.xlsx", sheet_name="Sheet1")
```

```python
df = pd.concat([df_2022, df_2023])
```

```python
df_grouped = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()
```

```python
# 商品と売上年でグループ化し、合計金額を集計
df_grouped = df.groupby(['商品', '売上年']).agg({'金額（千円）': 'sum'}).reset_index()

# 結果を表示
print(df_grouped)
```

```python
# 新規Excelファイルに書き込み
output_file = "売上集計表.xlsx"
df_grouped.to_excel(output_file, index=False)
print(f"結果が{output_file}に保存された")
```

```python
import pandas as pd
from openpyxl import load_workbook
```

```python
# エクセルファイルへの書き込み
with pd.ExcelWriter('売上集計表.xlsx', engine='openpyxl') as writer:
    df_grouped.to_excel(writer, sheet_name='sheet1', index=False)
    workbook = writer.book
    sheet = workbook.active
    # 列幅設定をここに移動
    sheet.column_dimensions['A'].width = 15
    sheet.column_dimensions['b'].width = 10
    sheet.column_dimensions['c'].width = 11
```

```python
wb = load_workbook(output_file)
ws = wb.active
```

```python
header_fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")
```

```python
for cell in ws[1]:
    cell.fill = header_fill
```

```python
wb.save(output_file)
```
